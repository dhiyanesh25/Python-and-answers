```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = {
    'x': [1, 2, 3, 4],
    'y': [15,20,24,32]
}

sns.lineplot(x='x', y='y', data=data)

plt.title('Basic Line Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

plt.show()
```

```python
data = {
    'CLASS': ['A', 'B', 'C', 'D'],
    'STUDENT PASSED': [8, 3, 10, 5]
}

# Create bar plot
sns.barplot(x='CLASS', y='STUDENT PASSED', data=data)

plt.title('NO OF STUDENTS PASSED IN EACH CLASS')
plt.xlabel('CLASS')
plt.ylabel('STUDENT PASSED')
plt.show()
```

```python
data = {
    'x': [1, 2, 3, 4, 5],
    'y': [20, 7, 15, 2, 4]
}

sns.scatterplot(x='x', y='y', data=data)
plt.title('Scatter Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

plt.show()
```
