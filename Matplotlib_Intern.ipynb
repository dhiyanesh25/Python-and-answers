PIE CHART


```python
import matplotlib.pyplot as plt
labels = ['A', 'B', 'C', 'D']
sizes = [25, 40, 25, 10]
colors = ['grey', 'yellow', 'blue', 'orange']
explode = (0, 0, 0, 0.1)  # explode the first slice

plt.figure(figsize=(6,6))
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%',
        shadow=True, startangle=50)
plt.title('Pie Chart')
plt.axis('equal')
plt.show()
```

HISTOGRAM

```python
import matplotlib.pyplot as plt
data = np.random.randn(1000)

plt.figure(figsize=(5,4))
plt.hist(data, bins=10, color='yellow', edgecolor='black', alpha=0.7)
plt.title('Histogram')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

QN3)BAR CHART FOR TOTAL PURCHASE IN A AGE GROUP

```python
import matplotlib.pyplot as plt
ages = ['18-28', '28-38', '38-58', 'Above 58']
values = [3, 7, 5, 9]

plt.figure(figsize=(4,3))
plt.bar(ages, values, color=['blue', 'green', 'red', 'yellow'])
plt.title('Bar Plot')
plt.xlabel('AGE')
plt.ylabel('PURCHASE')
plt.show()
```

Countour plot

```python
import matplotlib.pyplot as plt
x = np.linspace(-5, 5, 50)
y = np.linspace(-5, 5, 50)
X, Y = np.meshgrid(x, y)
Z = np.sin(np.sqrt(X**2 + Y**2))

plt.figure(figsize=(4, 3))
contour = plt.contour(X, Y, Z, 20, cmap='RdGy')
plt.colorbar(contour)
plt.title('Contour Plot Example')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```
