Replacing Null Values

```python
#Now we are filling the null values with zero and i have assigned to new Dataframe df2
df2=df.fillna(value=0)
df2
```

```python
#Now when we check the df2 there is no null values as we have filled it with Zero
df2.isnull().sum().sum()
```

```python
#Filling Null Values with Previous values
df3=df.fillna(method='pad')
df3
```

```python
df
```

```python
#Filling the null values with next values
df4=df.fillna(method='bfill')
df4
```

```python
#Filling null values with column wise
df5=df.fillna(method='pad', axis=1)
df5
```

```python
#This the same but it copies values from the right hand side
df6=df.fillna(method='bfill', axis=1)
df6
```

```python
#Now I have to fill the Null values from different column in different values 
df7=df.fillna({'society':'abcd',
               'balcony':'defg'})
df7
```

```python
#Filling the Null values with the mean of a column the mean() can also be replaced with min(),max() and median()
df8=df.fillna(value=df['balcony'].mean())
df8
```

Dropna()

```python
df
```

```python
#It drops all the Null values from the Dataset
df9=df.dropna()
df9
```

```python
#In how we can fill all or any 
#All is used if in a column there are more Null then it would remove the entire column
#Any is used if any one Values contains Null values
df10=df.dropna(how='all')
df10
```

Replace()

```python
#This replace() function is used to replace the NUll values with the required values
import numpy as np
df11=df.replace(to_replace=np.nan,value=875465)
df11
```

Interpolate()

```python
df['balcony']=df['balcony'].interpolate(method='linear')
df
```

```python
#In this we have added a limit direction to move forward
df['balcony']=df['balcony'].interpolate(method='linear',limit_direction='forward')
df
```
